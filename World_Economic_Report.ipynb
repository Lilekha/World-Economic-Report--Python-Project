# Objective 1: Prepare & QA the Data 
### Getting the data ready for analysis by importing and joining the data, creating new columns, and performing basic profiling and QA.

#### I. Import the World Bank Excel file.

```python
import pandas as pd
import numpy as np

world_bank = pd.read_excel("WorldBank.xlsx").rename({"Country Name": "Country"}, axis=1)

world_bank.head()
```

#### II. Use "GDP" and "GDP per Capita" to calculate population in millions and create a "Population (M)" column.

```python
world_bank["Population (M)"] = (world_bank["GDP (USD)"]/ world_bank["GDP per capita (USD)"] / 1_000_000)

world_bank.head()
```

#### III. Perform basic profiling & QA on the data.

```python
world_bank.info()
```

```python
world_bank.describe()
```

```python
hdi = pd.read_csv("hdi.csv")

hdi.head()
```

#### IV. Filter the data to 2014 and join it with the HDI csv file on "Country Code".

```python
hdi = pd.read_csv("hdi.csv")

wb_hdi_2014 = world_bank.query("Year == 2014").merge(hdi[["iso3", "hdi_2014"]], how="left", left_on="Country Code", right_on="iso3")

wb_hdi_2014.head()
```

```python
wb_hdi_2014.describe()
```

# Objective 2: Prepare the data for visualization
### Transforming and shaping the data into tables that will be used as the source for data visualizations.

#### I. Create a "gdp_pivot" table with years as rows, regions as columns, sum of GDP as values.

```python
gdp_pivot = world_bank.pivot_table(index="Year", columns="Region", values="GDP (USD)", aggfunc="sum")

gdp_pivot.head()
```

#### II. Create "pop_pivot" table with years as rows, regions as columns, and sum of population as values.

```python
pop_pivot = world_bank.pivot_table(index="Year", columns="Region", values="Population (M)", aggfunc="sum")

pop_pivot.head()
```

#### III. Create "wb_hdi_by_region" table by calculating the average HDI for each region and sorting from highest average HDI to lowest.

```python
wb_hdi_by_region = wb_hdi_2014.groupby("Region").agg({"hdi_2014": "mean"}).sort_values("hdi_2014", ascending=False)

wb_hdi_by_region.head()
```

# Objective 3: Visualize the data
### Creating 5 charts to visualize trends, relationships, and comparisons across regions

#### I. Create a stacked area chart showing the growth of GDP over time, with each stack representing a geographic region.

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

ax.stackplot(
gdp_pivot.index,
[gdp_pivot [region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index
)
ax. legend(loc="upper left")
ax.set_title("GDP has Grown Exponentially Over Time")
ax.set_ylabel("GDP (Trillions)")
```
